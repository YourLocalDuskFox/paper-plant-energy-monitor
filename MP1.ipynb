```python
#import libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import joblib
import sklearn
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Read, import, and copy the dataset
df = pd.read_csv('em_data.csv')
df1 = df.copy() 
```

```python
#Check for missing values
print(f"Check for missing values:\n{df1.isna().sum(),df1.shape}")
#there are none, so no need to fillna
```

```python
#encode categorical values into numbers
df1 = pd.get_dummies(df1,dtype=int)
print(f"Dataframe with categorical features encoded:\n{df1}")
```

```python
#Split labels & features into x and y
X = df1.drop(columns=["Energy (kWh)"])
y =df1['Energy (kWh)']
```

```python
#Split into training and testing data before creating algorithm
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python

model = RandomForestRegressor(
    n_estimators=200,      
    max_depth=10,          
    random_state=4)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error:{rmse}")
print(f"R²:{r2}")
```

```python
filename = f"MP1_{sklearn.__version__}.joblib"
joblib.dump(model, filename)
print(f"Model saved as {filename}")
```
